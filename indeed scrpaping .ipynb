```python
def extract_job_title_from_result(soup): 
    ReviewRating = pd.DataFrame()
    con=''
    pro=''
    for div in soup.find_all(name='div', attrs={'class':'cmp-Review-container'}):
        #print (div)  
        print('--------------------------------------------')  
        for a in div.find_all(name='div',attrs={'class':'cmp-ReviewRating-text'}):
            print(a.get_text())
        for title in div.find_all(name='div',attrs={'class':'cmp-Review-title'}):
            print(title.get_text())
        for author in div.find_all(name='div',attrs={'class':'cmp-Review-author'}):
            print((author.get_text()).split('-'))
        for text in div.find_all(name='div',attrs={'class':'cmp-Review-text'}):
            print(text.get_text()) 
        for p in div.find_all(name='div',attrs={'class':'cmp-ReviewProsCons'}):
            #print(p)
            print(len(p))
            if len(p)==2:
                print("found pro and cons")
                for pro in p.find_all(name='div',attrs={'class':'cmp-ReviewProsCons-prosText'}):
                    print("____pros_______")
                    print(pro.get_text())
                for con in p.find_all(name='div',attrs={'class':'cmp-ReviewProsCons-consText'}):
                    print("____cons_______")
                    print(con.get_text())
    return()
#extract_job_title_from_result(soup)
```
